```python
import time
modelos = ["VGG-Face", "Facenet", "OpenFace", "DeepFace", "DeepID", "ArcFace", "Dlib"]

# Rutas de las imágenes a comparar
img1 = "obama.jpg"
img2 = "obama2.jpg"

for modelo in modelos:
    try:
        print(f"\nProbando el modelo: {modelo}")
        inicio_tiempo = time.time()
        
        # Realiza la verificación con el modelo actual
        resultado = DeepFace.verify(img1_path=img1, img2_path=img2, model_name=modelo)
        tiempo_procesamiento = time.time() - inicio_tiempo
        
        print(f"Resultado para {modelo}: {resultado}")
        print(f"Tiempo de predicción: {tiempo_procesamiento:.2f} segundos")
    except Exception as e:
        print(f"Error al probar el modelo {modelo}: {e}")
```

```python
import cv2
import time
def resize_img(img):
  (h, w) = img.shape[:2]

  # Definir el nuevo ancho
  nuevo_ancho = 50

  # Calcular el ratio para mantener la relación de aspecto
  ratio = nuevo_ancho / float(w)
  nuevo_alto = int(h * ratio)

  # Redimensionar la imagen
  return cv2.resize(img, (nuevo_ancho, nuevo_alto))
```

```python
modelos = ["VGG-Face", "Facenet", "OpenFace", "DeepFace", "DeepID", "ArcFace", "Dlib"]

# Rutas de las imágenes a comparar
img1 = resize_img(cv2.imread("guerrero/guerrero1.png"))
img2 = resize_img(cv2.imread("guerrero/taruco.png"))

for modelo in modelos:
    try:
        print(f"\nProbando el modelo: {modelo}")
        inicio_tiempo = time.time()

        # Realiza la verificación con el modelo actual
        resultado = DeepFace.verify(img1_path=img1, img2_path=img2, model_name=modelo)
        tiempo_procesamiento = time.time() - inicio_tiempo

        print(f"Resultado para {modelo}: {resultado}")
        print(f"Tiempo de predicción: {tiempo_procesamiento:.2f} segundos")
    except Exception as e:
        print(f"Error al probar el modelo {modelo}: {e}")
```

```python
import time
img1 = resize_img(cv2.imread("guerrero/guerrero1.png"))
inicio_tiempo = time.time()
analyze_face = DeepFace.analyze(img1)
fin_tiempo = time.time()
print("")
print(f"Tiempo de ejecución: {fin_tiempo - inicio_tiempo} segundos")
import pprint
formatted_data = pprint.pformat(analyze_face, indent=2, width=80)
print(formatted_data)
```
